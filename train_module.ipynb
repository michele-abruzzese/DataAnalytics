Load the dataset

```python
import pandas as pd

FILENAME = "train.csv"
df = pd.read_csv(FILENAME)
Y = df.iloc[:,0]
X = df.iloc[:,1:]
```

```python
print(df)
```

```python
# Raggruppa per anno e conta il numero di canzoni
df_counts = df.groupby('Year').size().reset_index(name='qty')

# Ordina il DataFrame per anno
df_counts = df_counts.sort_values(by='Year')

# Visualizza il nuovo DataFrame
print(df_counts)

import matplotlib.pyplot as plt
# Crea l'istogramma
plt.figure(figsize=(12, 6))
plt.bar(df_counts['Year'], df_counts['qty'], edgecolor='black')
plt.title('Distribuzione delle Canzoni per Anno')
plt.xlabel('Anno')
plt.ylabel('Numero di Canzoni')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostra l'istogramma
plt.show()
```

```python
import seaborn as sns

# Sostituisci 'NOME_COLONNA' con il nome effettivo della colonna che vuoi esplorare
column_to_plot = 'S0'

# Calcola il valore medio per ogni anno
df_mean = df.groupby('Year')[column_to_plot].mean().reset_index()

# Crea un grafico a dispersione o una linea
plt.figure(figsize=(12, 6))
#sns.scatterplot(x='Year', y=column_to_plot, data=df, edgecolor='black', alpha=0.7)
# Oppure, utilizza un grafico a linea con i valori medi:
sns.lineplot(x='Year', y=column_to_plot, data=df_mean, color='red', label=f'Media di {column_to_plot}')

plt.title(f'Andamento di {column_to_plot} e Media Annuale')
plt.xlabel('Anno')
plt.ylabel(column_to_plot)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostra il grafico
plt.show()
```

```python
# Seleziona le colonne per la matrice di correlazione
columns_to_correlate = ['S' + str(i) for i in range(90)]  # Genera nomi delle colonne da 'S0' a 'S89'

# Crea la matrice di correlazione
correlation_matrix = df[columns_to_correlate].corr()

# Visualizza la matrice di correlazione utilizzando seaborn
plt.figure(figsize=(18, 16))
sns.heatmap(correlation_matrix, cmap='coolwarm', linewidths=.5)
plt.title('Matrice di Correlazione tra le colonne S0-S89')
plt.show()

# Trova colonne con correlazione maggiore di 0.7
highly_correlated_columns = set()
for col in correlation_matrix.columns:
    correlated_columns = correlation_matrix.index[correlation_matrix[col] > 0.7].tolist()
    correlated_columns.remove(col)  # Rimuovi la stessa colonna dalla lista
    highly_correlated_columns.update(correlated_columns)

print("Insieme di colonne con correlazione maggiore di 0.7:")
print(highly_correlated_columns)
```
